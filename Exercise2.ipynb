Write a python program to consider some values as (x,y) co-ordinate values and plot the graph using a line graph. The color of the line graph should be red. 

```python
# !python -m pip install matplotlib

import matplotlib.pyplot as plt

# Sample x, y coordinate values
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 6, 8, 10]

# Plot the line graph with red color
plt.plot(x_values, y_values, color='red', marker='o', linestyle='-')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')

# Show the plot
plt.show()
```
